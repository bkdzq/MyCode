```python
#  牛顿法 newton's method 求根
import matplotlib.pyplot as plt
def generate_newton_iters(x0, number):
    iterates = [x0]
    errors = [abs(x0 - 1.)]
    for i in range(number):
        x0 = x0 - (x0*x0 - 1)/(2*x0)
        iterates.append(x0)
        errors.append(abs(x0 - 1.))
    return iterates, errors
iterates, errors = generate_newton_iters(2.0, 5)
fig, (ax1, ax2) = plt.subplots(1, 2, tight_layout=True) # 1 row, 2columns
#fig.dpi = 600  #  图像质量不高，可以这里调，然后截图
ax1.plot(iterates, "x")
ax1.set_title("Iterates")
ax1.set_xlabel("i")
ax1.set_ylabel("x_i")

ax2.semilogy(errors, "x") # plot y on logarithmic scale
ax2.set_title("Error")
ax2.set_xlabel("i")
ax2.set_ylabel("Error")
```

```python
# 为下一代码块创造数据   迭代与误差
def generate_newton_iters(x0, number):
       iterates = [x0]
       errors = [abs(x0 - 1.)]
       for _ in range(number):
           x0 = x0 - (x0*x0 - 1.)/(2*x0)
           iterates.append(x0)
           errors.append(abs(x0 - 1.))
       return iterates, errors
iterates, errors = generate_newton_iters(2.0, 5)
print("iterates is: ",iterates,'\n',"error is: ", errors)
```

```python
# 加一个color map    面：z = np.cos(2*np.pi*t)*np.exp(-t) 
# in this case, the larger the z value, the lighter the shade of gray
X = np.linspace(-2, 2)
Y = np.linspace(-2, 2)
x, y = np.meshgrid(X, Y)
t = x**2 + y**2  # small efficiency
z = np.cos(2*np.pi*t)*np.exp(-t)

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.plot_surface(x, y, z, cmap="binary_r")  # 这里加colormap

ax.set_title("Surface with colormap")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")
```

```python
#Python – Legendre polynomials using Recursion relation
# Legendre polynomial
def P(n, x): 
    if(n == 0):
        return 1 # P0 = 1
    elif(n == 1):
        return x # P1 = x
    else:
        return (((2 * n)-1)*x * P(n-1, x)-(n-1)*P(n-2, x))/float(n)

  
# This is for use in webbrowser, can be ignored.
#matplotlib.use('Agg') 
  
import matplotlib.pyplot as plt
import numpy as np
#matplotlib.use('TkAgg')#  加上这句话才能运行
  
# Creating an array of x values
x = np.linspace(-1, 1, 200) 
  
# for which polynomial values are evaluated and plotted
fig, ax = plt.subplots()
for i in range(1, 5):
  
    # Labelling according to order
    ax.plot(x, P(i, x), label ="P"+str(i)) 
  
ax.legend(loc ="best")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()



```
